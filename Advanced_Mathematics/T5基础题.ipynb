求函数$f(x)=sin\frac{\pi x}{2(1+x^2)}$的值域。

# 解：

$$ 
\begin{align}
\begin{cases}
(x-1)^2 \geq 0 & \\
(x+1)^2 \geq 0 &
\end{cases}
& \Rightarrow 
\begin{cases}
2x \leq x^2 + 1 & \\
2x \geq -(x^2 + 1) &
\end{cases} \\
& \Rightarrow 
-(x^2 + 1) \leq 2x \leq (x^2 + 1) \\
& \Rightarrow
-1 \leq \frac{2x}{x^2+1} \leq 1 \\
& \Rightarrow
-\frac{\pi}{4} \leq \frac{\pi}{4}\frac{2x}{x^2+1} \leq \frac{\pi}{4} \\
& \Rightarrow
sin-\frac{\pi}{4} \leq sin\frac{\pi x}{2(1+x^2)} \leq sin\frac{\pi}{4} \\ 
& \Rightarrow
-1 \leq sin\frac{\pi x}{2(1+x^2)} \leq 1
\end{align}
$$

# 答:
$f(x)$的值域是$[-1, 1]$.
