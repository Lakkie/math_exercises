已知$f(x)=arcsin\frac{4}{x}$，求定义域$D_f$及$f(-4),f(\frac{8}{\sqrt{2}}),f(\frac{8}{\sqrt{3}})$.

# 解：
$$
-1 \leq \frac{4}{x} \leq 1
\Rightarrow
x \leq -4 \quad or \quad x \geq 4
$$

# 答：
$$
D_f = (-\infty, -4] \cup [4, +\infty)
$$

$$
f(-4) = arcsin(-1) = -\frac{\pi}{2}
$$

$$
f(\frac{8}{\sqrt{2}}) = arcsin(\frac{\sqrt{2}}{2}) = \frac{\pi}{4}
$$

$$
f(\frac{8}{\sqrt{3}}) = arcsin(\frac{\sqrt{3}}{2}) = \frac{\pi}{3}
$$
