```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-1, 1, 100)  # 从-1到1生成100个点
y_arcsin = np.arcsin(x)  # 反正弦函数
y_arccos = np.arccos(x)  # 反余弦函数
y_arctan = np.arctan(x)  # 反正切函数
y_arccot = np.pi / 2 - np.arctan(x) # 反余切函数

plt.figure(figsize=(10, 5))

# 绘制反正弦函数图像
plt.subplot(2, 2, 1)
plt.plot(x, y_arcsin, label='arcsin(x)', color='g')
plt.title('arcsin(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
# 绘制反余弦函数图像
plt.subplot(2, 2, 2)
plt.plot(x, y_arccos, label='arccos(x)', color='r')
plt.title('arccos(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()

# 绘制反正切函数图像
plt.subplot(2, 2, 3)
plt.plot(x, y_arctan, label='arctan(x)', color='g')
plt.title('arctan(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()

# 绘制反余切函数图像
plt.subplot(2, 2, 4)
plt.plot(x, y_arccot, label='arccot(x)', color='r')
plt.title('arccot(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
