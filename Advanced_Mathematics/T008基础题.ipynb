已知$ f(x)=
\begin{cases}
\frac{x}{\pi}|sinx|, |x| < \frac{\pi}{3} & \\
0, |x| \geq \frac{\pi}{3} & 
\end{cases}
$，求$f(\frac{\pi}{6}), f(-\frac{\pi}{4}), f(2)$.

# 解：

$$
\begin{align}
f(\frac{\pi}{6})
&= \frac{\pi}{6} \cdot \frac{1}{\pi} \cdot |sin\frac{\pi}{6}| \\
&= \frac{1}{6} \cdot \frac{1}{2} \\
&= \frac{1}{12} \\
f(-\frac{\pi}{4})
&= -\frac{\pi}{4} \cdot \frac{1}{\pi} \cdot |sin(-\frac{\pi}{4})| \\
&= -\frac{1}{4} \cdot \frac{\sqrt{2}}{2} \\
&= -\frac{\sqrt{2}}{8} \\
f(2) &= 0
\end{align}
$$
