已知$f(x)=arccos(2x+1)+log_2(3x+2)$，求定义域$D_f$.

# 解：
$$
\begin{cases}
-1 \leq 2x + 1 \leq 1 & \\
3x + 2 > 0 & 
\end{cases}
\Rightarrow
\begin{cases}
-1 \leq x \leq 0 & \\
x > -\frac{2}{3} & 
\end{cases}
\Rightarrow
-\frac{2}{3} < x \leq 0
$$

# 答：
函数$f(x)$的定义域为$ (-\frac{2}{3}, 0].$
